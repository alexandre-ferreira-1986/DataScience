O coronavírus (COVID-19) é uma doença infecciosa causada pelo vírus SARS-CoV-2.

A maioria das pessoas infectadas pelo COVID-19 apresentam sintomas leves e moderados e se recuperam sem tratamento especial.

No entanto, algumas desenvolvem um quadro grave e precisam de atendimento médico.

O intuito deste artigo é realizar uma análise a partir dos dados fornecidos neste [repositório público](https://github.com/owid/covid-19-data/tree/master/public/data). 

Desde Janeiro de 2020 a  OMS declarou o surto de coronavírus e desde então há muitas informações espalhadas e cada hora uma nova variante surge elevando o grau de vigilância.

Em Março de 2020 a OMS caracterizou o COVID-19 como uma **pandemia**. A termologia utilizada se refere à distribuição geográfica da doença e não necessariamente com relação a sua gravidade.

```
# Construindo gráfico básico
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize=(16,8))

top5_death = df.loc[df.date == '2021-09-20', ['location', 'total_deaths']].sort_values(by="total_deaths", ascending=False)[0:6]
ax[0].bar('location', 'total_deaths', data=top5_death)
ax[0].set_title("Cinco paises com mais mortes de COVID-19 \n **SEM AJUSTE NO GRÁFICO**")


#Construindo gráfico mais completo
bars = ax[1].bar(
    x = 'location',
    height = 'total_deaths',
    data = top5_death,
    color = "#650000"
)

# Matplotlib: Spines are the lines connecting the axis tick marks and noting the boundaries of the data area.
ax[1].spines['top'].set_visible(False)
ax[1].spines['right'].set_visible(False)
ax[1].spines['left'].set_visible(False)
ax[1].spines['bottom'].set_color('#650000')

ax[1].tick_params(bottom=False, left=False)
ax[1].set_axisbelow(True)
ax[1].yaxis.grid(True, color='#EEEEEE')
ax[1].xaxis.grid(False)
ax[1].ticklabel_format(axis = 'y', style = 'plain')
plt.setp(ax[1].get_xticklabels(), rotation=30, ha = 'right')
ax[1].set_title("Cinco paises com mais mortes de COVID-19 \n **COM AJUSTE NO GRÁFICO**")

# Pegar a cor das barras para que o texto tenha a mesma cor
bar_color = bars[0].get_facecolor()

# Adicionar anotações ao topo das barras.
for bar in bars:
  ax[1].text( 
      bar.get_x() + bar.get_width() / 2,
      bar.get_height() + 5000,
      round(bar.get_height()),
      horizontalalignment='center',
      color=bar_color,
      weight='bold',
      fontsize = 11
  )



fig.tight_layout(pad = 10.0)
```

```
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize=(16,8))

top5_death = df.loc[df.date == '2021-09-20', ['location', 'total_cases']].sort_values(by="total_cases", ascending=False)[0:6]
ax[0].bar('location', 'total_cases', data=top5_death)
ax[0].set_title("Cinco paises com mais casos de COVID-19 \n **SEM AJUSTE NO GRÁFICO**")


bars = ax[1].bar(
    x = 'location',
    height = 'total_cases',
    data = top5_death,
    color = "#152D35"
)

# Matplotlib: Spines are the lines connecting the axis tick marks and noting the boundaries of the data area.
ax[1].spines['top'].set_visible(False)
ax[1].spines['right'].set_visible(False)
ax[1].spines['left'].set_visible(False)
ax[1].spines['bottom'].set_color('#152D35')

ax[1].tick_params(bottom=False, left=False)
ax[1].set_axisbelow(True)
ax[1].yaxis.grid(True, color='#EEEEEE')
ax[1].xaxis.grid(False)
ax[1].ticklabel_format(axis = 'y', style = 'plain')
plt.setp(ax[1].get_xticklabels(), rotation=30, ha = 'right')
ax[1].set_title("Cinco paises com mais casos de COVID-19 \n **COM AJUSTE NO GRÁFICO**")
# Grab the color of the bars so we can make the
# text the same color.
bar_color = bars[0].get_facecolor()

# Add text annotations to the top of the bars.
# Note, you'll have to adjust this slightly (the 0.3)
# with different data.
for bar in bars:
  ax[1].text( 
      bar.get_x() + bar.get_width() / 2,
      bar.get_height() + 500000,
      round(bar.get_height()/1000000),
      horizontalalignment='center',
      color=bar_color,
      weight='bold',
      fontsize = 10
  )

ax[1].text(1.8, 37000000, 'Em Milhões de casos', fontsize=14, color = "#345B63")
ax[1].annotate("",
            xy=(1.1, 34500000), xycoords='data',
            xytext=(1.8, 37000000), textcoords='data',
            arrowprops=dict(arrowstyle="fancy",
                            connectionstyle="arc3, rad = -0.2")
            )

fig.tight_layout(pad = 10.0)
```
